```python
import openai
from jinja2 import Environment
import json
import yaml
```

```python
from config import API_KEY
openai.api_key = API_KEY

# Sends prompt to ChatGPT, returns response
def getGPTResponse(prompt: str):
    messages = [ {"role": "system", "content": prompt} ]
    chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-16k",
        temperature=0,
        messages=messages
    )
    reply = chat.choices[0].message.content
    return reply
```

```python
template = Environment().from_string(promptTemplate)
```

```python
prompts = [ template.render({'exerciseName': exercise }) for exercise in exerciseList ]
```

```python
responses = {}
```

```python
for i in range(260, len(prompts)):
    responses[exerciseList[i]] = getGPTResponse(prompts[i])
```

```python
with open(F'./responses.json', 'w') as f:
        json.dump(responses, f)
with open(F'./responses.yaml', 'w') as f:
        yaml.dump(responses, f)
```
